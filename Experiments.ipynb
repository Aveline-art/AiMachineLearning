```python
import numpy as np
import matplotlib.pyplot as plt
from experiment import *
```

```python
def plot_results(alpha, epsilon, num_trainings, num_batches, num_test_games):

    ai = HogAI(q=dict(), alpha=alpha, epsilon=epsilon)

    x, y = trials(train_function(ai), num_trainings, num_batches, test_function(strat_rand, num_test_games), say=speak)
    x = np.array(x)
    y = np.array(y)
    fit = np.polyfit(x, y, 2)

    plt.plot(x, y, label="win rate")
    plt.plot(x, np.polyval(fit, x), "r", label="best fit")

    # configuring the x axis
    plt.xlabel('number of training games')
    plt.xlim(0,num_trainings)
    # configuring the y axis
    plt.ylabel('win rate')
    plt.ylim(0.3, 0.7)

    # giving a title to my graph
    plt.title(f'Win rate alpha {alpha} , epsilon {epsilon}')

    # adding a legend
    plt.legend()

    # function to show the plot
    plt.show()
```

```python
num_trainings = 10000
num_batches = 100
num_test_games = 1000
```

```python
alpha = 0.00
epsilon = 0.1

plot_results(alpha, epsilon, num_trainings, num_batches, num_test_games)
```
